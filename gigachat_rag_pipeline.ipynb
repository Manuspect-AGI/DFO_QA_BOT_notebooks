```python
from langchain.schema import HumanMessage, SystemMessage
from langchain.chat_models.gigachat import GigaChat

# инициализируем гигачат

credential = '' # тут ввести api key gigachat
# model = 'GigaChat Lite' #название модели

chat = GigaChat(credentials=credential, verify_ssl_certs=False, temperature=0.01, max_tokens = 8192,)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# можно менять чанк сайз и чанк оверлап для повышения качества
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1024, chunk_overlap=64)
all_splits = text_splitter.split_documents(pages)
```

```python
all_splits
```

Векторизуем чанки и кладем в хрому

```python
from langchain_community.vectorstores import Chroma
import chromadb

# client = chromadb.HttpClient(host="51.250.111.105", port=8040)
vectorstore = Chroma.from_documents(embedding=embeddings, documents=all_splits)
# за че search type отвечает?
retriever = vectorstore.as_retriever(search_kwargs={"k": 6})
```

Тестим ретривер

```python
problem = retriever.invoke('Проблематика')
ans = []
for i in problem:
    ans.append(' '.join(i.page_content.split('\n')))

ans
```

```python
problem = retriever.invoke('Какой стек технологий надо использовать в хакатоне?')
ans = []
for i in problem:
    ans.append(' '.join(i.page_content.split('\n')))

ans
```

```python
sys_message = 'ТЫ помощник по хакатону который отвечает на вопросы по нему' # системный промпт

messages_1 = [
    SystemMessage(
        content=sys_message
    )
]

question = 'Какой стек технологий надо использовать в хакатоне?'
context = '.'.join(x for x in ans)

def get_context(question):
    problem = retriever.invoke(question)
    for i in problem:
        ans.append(' '.join(i.page_content.split('\n')))
    return '.'.join(x for x in ans)

context = get_context(question)

prompt = f'Ответь на вопрос: {question}. Используй контекст: {context}. Если в контексте нет ответа на вопрос - пиши не знаю.'

messages_1.append(HumanMessage(content=prompt))
res = chat(messages_1)
print(res.content)
```
